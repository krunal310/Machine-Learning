```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```
dataset = pd.read_csv('/content/drive/MyDrive/8th Sem/Machine learning/Exp 4: Polynomial Regression/housing.csv')
dataset.columns
```

```
x = dataset[['bathrooms','sqft_living','grade','sqft_above','sqft_living15']]
y = dataset['price']
```

```
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 1/3, random_state = 0 )
```

```
from sklearn.preprocessing import PolynomialFeatures
p_reg = PolynomialFeatures(degree=2)
```

```
xtrain_2d = p_reg.fit_transform(xtrain)
```

```
from sklearn.linear_model import LinearRegression
l_reg = LinearRegression()
l_reg.fit(xtrain_2d,ytrain)
```

```
ytrain_pred = l_reg.predict(xtrain_2d)
ytest_pred = l_reg.predict(p_reg.fit_transform(xtest))
```

```
from sklearn import metrics

print('Evalution of Train set\n')
print('MAE:', metrics.mean_absolute_error(ytrain, ytrain_pred))
print('MSE:', metrics.mean_squared_error(ytrain, ytrain_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytrain, ytrain_pred)))
print('R2 Score: ', metrics.r2_score(ytrain, ytrain_pred))
```

```
from sklearn import metrics

print('Evalution of Test set\n')
print('MAE:', metrics.mean_absolute_error(ytest, ytest_pred))
print('MSE:', metrics.mean_squared_error(ytest, ytest_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, ytest_pred)))
print('R2 Score: ', metrics.r2_score(ytest, ytest_pred))
```
